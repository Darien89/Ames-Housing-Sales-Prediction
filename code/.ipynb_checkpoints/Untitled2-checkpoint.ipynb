```python
features = [col for col in train._get_numeric_data().columns if col != ['saleprice']]
```

```python
features
```

```python
#create X and y
X = train[features]
y = train['saleprice']

#verify dimensions
print('X: ', X.shape)
print('y: ', y.shape)
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=42)
```

```python
#instantiate standard scaler
ss = StandardScaler()
#fit standard scaler to X_train and transform 
X_train_sc = ss.fit_transform(X_train)
#fit standard scaler to X_test and transform
X_test_sc = ss.transform(X_test)

#standardize test set data 
test_sc = ss.fit_transform(test)
```

```python
lasso_model = LassoCV(n_alphas=500, cv=10, verbose=1)
lasso_model.fit(X_train_sc, y_train)

print (lasso_model.alpha_)
```

```python
lasso = Lasso(alpha=lasso_model.alpha_).fit(X_train_sc, y_train)

lasso_scores = cross_val_score(lasso, X_train_sc, y_train, cv=10)

print (lasso_scores)
```

```python
#generate predictions
lasso_predicition = lasso_model.predict(X_test_sc)
```

```python
#MSE
print(f'MSE: {mean_squared_error(y_test, lasso_predicition)}')
#RMSE
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, lasso_predicition))}')
#R^2
print(f'R^2: {r2_score(y_test, lasso_predicition)}')
```
